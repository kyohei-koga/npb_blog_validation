# 選手成績のクローリング
- 選手の成績情報をクローリングする。

```python
import pandas as pd
import sqlite3
import numpy as np
import scipy
import matplotlib.pyplot as plt
import seaborn as sns
import requests
import lxml.html
import time
%matplotlib inline
```

```python
# 更新の時はOBか否かで判断する。出来上がったdfから当該年度だけ抜き出すということになる。
# biosについて、新しい選手が入ると更新されるはずだからURlと照合してなければ取得ということになる。
# あいうえお索引ページ
response = requests.get('http://npb.jp/bis/players/all/index.html')
root = lxml.html.fromstring(response.content)
# 絶対URL化
root.make_links_absolute(response.url)
# 各索引のURLの取得
urls = root.cssselect('#pl_lk_unit a')

# 選手情報の取得
bios = []
pitchers = []
fielders = []
for i, url in enumerate(urls):
    # 選手一覧ページの取得
    response = requests.get(url.get('href'))
    root = lxml.html.fromstring(response.content)
    # 絶対URL化
    root.make_links_absolute(response.url)
    player_links = root.cssselect('.three_column_player > a')
    # 選手情報の取得
    bio, pitcher, fielder = get_player_info(player_links)
    
    bios.append(bio)
    pitchers.append(pitcher)
    fielders.append(fielder)


# データの結合
all_player_bios = pd.concat(bios, axis=0, ignore_index=True)
all_pitcher_score = pd.concat(pitchers, axis=0, ignore_index=True)
all_fielder_score = pd.concat(fielders, axis=0, ignore_index=True)
```

```python
# 選手の基本情報と成績の取得
def get_player_info(urls):
    columns = ['安打', '本塁打', '四球', '死球', '三振', '暴投', 'ボーク', '失点', '自責点', '防御率']
    player_bios = []
    fielder_score = []
    pitcher_score = []

    # 頭文字のリンクから取得
    for i, row in enumerate(urls):
        time.sleep(1)
        url = row.get('href')
        
        # インターネットの接続性でエラーが出ることがある。
        # やり直し処理で5回までやり直す。
        retries = 0
        max_retries = 5
        while True:
            try:
                response = requests.get(url)
                break
            except:
                retries += 1
                if retries>=max_retries:
                    raise Exception('too many retries in requests.')
                wait = 2**(retries - 1)
                time.sleep(wait)

        root = lxml.html.fromstring(response.content)
        # 選手名の取得
        name = root.cssselect('li#pc_v_name')[0].text_content().strip()
        print(name)
        # 基本情報、成績は表形式になっているので、pandasで直接取りに行く
        # インターネットの接続性でエラーが出ることがある。
        # やり直し処理で5回までやり直す。
        retries = 0
        max_retries = 5
        while True:
            try:
                fetched_df = pd.io.html.read_html(url)  # 表をDataframeの形で取得
                break
            except:
                retries += 1
                if retries>=max_retries:
                    raise Exception('too many retries in pandas.')
                wait = 2**(retries - 1)
                time.sleep(wait)       

        bios = fetched_df[0].set_index(0).T
        bios['url'] = url
        bios['name'] = name
        # 成績のデータはない場合もあるので、try文
        try:
            score = fetched_df[1]
            score['url'] = url
            score['name'] = name
            # 投手と打者では成績として表示されるデータが異なるので別々に保存
            if root.cssselect('.tab_btn.current')[0].text_content().strip()=='投手成績':
                # 投手の投球回の表示の仕方が微妙でうまく取得的ない。
                # pandasからデータの修正をする。

                # 登板、もしくは、年度情報がnanの場合は必要ない
                score.dropna(subset=['登板','年度'], axis=0, inplace=True)
                # 投球回数のせいでデータがずれるのでその修正
                score.loc[:, columns] = score.shift(-2, axis=1).loc[:, columns]
                # 必要なくなった列の削除
                score.drop(columns=['Unnamed: 24', 'Unnamed: 25'], inplace=True)
                pitcher_score.append(score)
            else:
                score.dropna(subset=['年度'], axis=0, inplace=True)
                fielder_score.append(score)
        except:
            pass
        player_bios.append(bios)
    
    player_bios_con = pd.concat(player_bios, axis=0, ignore_index=True)
    player_pitcher_con = pd.concat(pitcher_score, axis=0, ignore_index=True)
    player_fielder_con = pd.concat(fielder_score, axis=0, ignore_index=True)
    
    return player_bios_con, player_pitcher_con, player_fielder_con
```

```python
# sqlite3に保存
conn = sqlite3.connect('players_info.db')
c = conn.cursor()

# DFから直接書き込み。同じ名前のtableがあった場合は上書きする。
all_player_bios.to_sql('bios', conn, if_exists='replace')
all_pitcher_score.to_sql('pitcher_score', conn, if_exists='replace')
all_fielder_score.to_sql('fielder_score', conn, if_exists='replace')
```

```python
# 上書き
conn.commit()
```
